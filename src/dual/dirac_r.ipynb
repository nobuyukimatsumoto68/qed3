```python
# k1 = np.loadtxt("ev_k1free.dat")
# k2 = np.loadtxt("ev_k2free.dat")
# k4 = np.loadtxt("ev_k4free.dat")
# k8 = np.loadtxt("ev_k8free.dat")
# k16 = np.loadtxt("ev_k16free.dat")

k1 = np.loadtxt("dirac_r_1.dat")
k2 = np.loadtxt("dirac_r_2.dat")
k4 = np.loadtxt("dirac_r_4.dat")
k8 = np.loadtxt("dirac_r_8.dat")
k16 = np.loadtxt("dirac_r_16.dat")
# k32 = np.loadtxt("dirac_r_32.dat")
```

```python
plt.plot(k4.T[0], k4.T[1], marker="x", ls="none")
```

```python
plt.plot(k8.T[0], k8.T[1], marker="x", ls="none")
```

```python
plt.plot(k16.T[0], k16.T[1], marker="x", ls="none")
```

```python
k1s = k1[k1[:,0].argsort()]
k2s = k2[k2[:,0].argsort()]
k4s = k4[k4[:,0].argsort()]
k8s = k8[k8[:,0].argsort()]
k16s = k16[k16[:,0].argsort()]
```

```python
ev1 = np.array([
    [k1s[0][0], k1s[0][1]],
    [k2s[0][0], k2s[0][1]],
    [k4s[0][0], k4s[0][1]],
    [k8s[0][0], k8s[0][1]],
    [k16s[0][0], k16s[0][1]],
])
```

```python
k1
```

```python
ev1
```

```python
ks = np.array([1, 2, 4, 8, 16])[:ev1.shape[0]]
```

```python
plt.plot( 1.0/ks, np.abs(ev1.T[0]),
        marker="x",  label="spec $D_{{\\rm ov}}$")

plt.plot( 1.0/ks, 0.02/ks**2, ls="dashed", label="quad scaling" )

plt.xscale("log")
plt.yscale("log")

plt.xlabel("$1/L$")
plt.ylabel("${\\rm Re}\, \\lambda_{1/2}$")

plt.tight_layout()
plt.legend()

# plt.savefig("figs/Re.pdf", bbox_inches='tight')
```

```python
plt.plot( 1.0/ks, np.abs(ev1.T[1]),
        marker="x", label="spec $D_{{\\rm ov}}$")

plt.plot( 1.0/ks, 0.2/ks, ls="dashed", label="linear scaling" )

plt.xscale("log")
plt.yscale("log")

plt.xlabel("$1/L$")
plt.ylabel("${\\rm Im}\, \\lambda_{1/2}$")

plt.tight_layout()
plt.legend()

# plt.savefig("figs/Im.pdf", bbox_inches='tight')
```
