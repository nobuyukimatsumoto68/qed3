```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_refine=5
```

```python
ev = np.loadtxt( "evd_n"+str(n_refine)+".dat" )
plt.scatter( ev.T[0], ev.T[1], s=2.0, label="n_refine="+str(n_refine) )

plt.xlabel("$\\lambda_R$")
plt.ylabel("$\\lambda_I$")

plt.legend()

plt.savefig("ev_dual_scatter"+str(n_refine)+".pdf")
```
