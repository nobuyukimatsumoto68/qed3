```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=16
```

```python
ev = np.loadtxt( "evf"+str(n)+".dat" )
evc = np.loadtxt( "evf"+str(n)+"_cu.dat" )
# ev2 = np.loadtxt( "ev.dat" )
```

```python
# ev = np.sort( ev, axis=0 )
# ev2 = np.sort( ev2, axis=0 )
```

```python
# plt.scatter( ev2.T[0], ev2.T[1] )
plt.scatter( ev.T[0], ev.T[1], label="$L="+str(n)+"$", s=5.0 )
plt.scatter( evc.T[0], evc.T[1], label="$L="+str(n)+"$", s=5.0 )

plt.xlim(0.0, 3.5)
plt.ylim(-3*np.pi/5, 3*np.pi/5)

plt.axis('equal')

plt.xlabel("$\\lambda_R$")
plt.ylabel("$\\lambda_I$")

plt.legend()

plt.savefig("evf_scatter"+str(n)+".pdf")
```
