```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for n_refine in np.arange(1,5):
res=[]
for n_refine in np.arange(1,4,1):

    ev = np.loadtxt( "ev_n"+str(n_refine)+".dat" )
    b = np.argsort( ev.T[0] )
    res.append([ n_refine, ev[b] ])
    plt.scatter( ev.T[0], ev.T[1], s=2.0, label="n="+str(n_refine) )
# plt.scatter( ev.T[0], -ev.T[1], s=1.0 )

# ev = np.loadtxt( "evd_n"+str(n_refine)+"_2.dat" )
# plt.scatter( ev.T[0], ev.T[1], s=2.0, label="n_refine="+str(n_refine) )

# plt.xlim(0.0, 3.5)
# plt.ylim(-3*np.pi/5, 3*np.pi/5)
# plt.ylim(-0.1, 0.1)

plt.axis('equal')

plt.xlabel("$\\lambda_R$")
plt.ylabel("$\\lambda_I$")

plt.legend(loc='lower right')

# plt.savefig("ev_dual_scatter"+str(n_refine)+".pdf")
plt.savefig("ev_scatter.pdf")
```

```python
for dat in res:
    ev = dat[1]
    n = dat[0]

    # b = np.argsort( ev.T[0] )
    plt.plot( np.abs(ev[::2].T[1])/ np.abs(ev[0][1]), marker="x", label="n="+str(n) )

plt.hlines(np.arange(1,6), 0, 30, label='', color='gray', linestyle=':')
plt.legend()

plt.xlim( 0, 30 )
plt.ylim( 0, 6 )
```
