```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ev = np.loadtxt( "ev_tim.dat" )
ev2 = np.loadtxt( "ev_tim2.dat" )
ev3 = np.loadtxt( "ev_tim3.dat" )

plt.scatter( ev.T[0], ev.T[1],
            marker='x', 
            label="Tim's spin structure" ) # , label="n_refine="+str(n_refine)

plt.scatter( ev2.T[0], ev2.T[1], 
            marker='.', 
            label="from geodesics (multiple patch, coord sing always outside patch)") # , label="n_refine="+str(n_refine)

plt.scatter( ev3.T[0], ev3.T[1], 
            marker='+', 
            label="from geodesics (single patch, coord sing avoided but exist)") # , label="n_refine="+str(n_refine)

plt.legend( loc='lower right' )
plt.savefig("comparison.pdf", bbox_inches="tight")
```
