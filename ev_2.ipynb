```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for n_refine in np.arange(1,5):
res=[]
for n_refine in np.arange(1,9,1):

    ev = np.loadtxt( "ev_n"+str(n_refine)+".dat" )
    b = np.argsort( ev.T[0] )
    res.append([ n_refine, ev[b] ])
```

```python
for dat in res[1::2]:
    ev = dat[1]
    n = dat[0]

    # b = np.argsort( ev.T[0] )
    plt.scatter( ev.T[0], ev.T[1], s=2.0, label="n="+str(n) )

# ev = np.loadtxt( "evd_n"+str(n_refine)+"_2.dat" )
# plt.scatter( ev.T[0], ev.T[1], s=2.0, label="n_refine="+str(n_refine) )

# plt.xlim(0.0, 3.5)
# plt.ylim(-3*np.pi/5, 3*np.pi/5)
# plt.ylim(-0.1, 0.1)

# plt.axis('equal')

plt.xlabel("${\\rm Re}\\lambda$")
plt.ylabel("${\\rm Im}\\lambda_I$")

plt.legend(loc='lower right')
plt.title("spectrum in complex plane (simplicial)")

plt.savefig("ev_scatter.pdf")
```

```python
for dat in res[:4]:
    ev = dat[1]
    n = dat[0]

    # b = np.argsort( ev.T[0] )
    plt.plot( np.abs(ev[::2].T[1])/ np.abs(ev[0][1]), marker="x", label="n="+str(n) )

plt.hlines(np.arange(1,6), 0, 30, label='', color='gray', linestyle=':')
plt.legend()

plt.xlim( 0, 30 )
plt.ylim( 0, 6 )

plt.title("${\\rm Im}\\lambda$ and deneracy structure (simplicial)")
plt.xlabel("positive eigenvalues")
plt.ylabel("${\\rm Im}\\lambda$")

plt.savefig("degen.pdf")
```

```python
nmax=7
spec2nd = np.array( [ [res[ii][0], np.abs(res[ii][1][5][1] / res[ii][1][0][1])] for ii in range(nmax) ] )
```

```python
plt.plot( 1.0/spec2nd.T[0], np.abs( spec2nd.T[1]-2.0 ), marker='x', 
         label='$\\vert {\\rm Re}\\lambda_{j=3/2}/{\\rm Re}\\lambda_{j=1/2}-2 \\vert$' )

plt.plot( 1.0/spec2nd.T[0], 0.1/spec2nd.T[0], ls='dashed', label="$1/n$")
plt.plot( 1.0/spec2nd.T[0], 0.5/spec2nd.T[0]**2, ls='dashed', label="$1/n^2$" )

plt.xscale("log")
plt.yscale("log")

plt.legend()
plt.title("$n$-dependence of ${\\rm Im}\\lambda$ (simplicial)")
plt.xlabel("$1/n$ (refinement level)")

plt.savefig("Imlambda_n.pdf")
```

```python
real2nd = np.array( [ [res[ii][0], res[ii][1][6][0]] for ii in range(len(res)) ] )

plt.plot( 1.0/real2nd.T[0], real2nd.T[1], marker='x', label='$j=3/2$' )
plt.plot( 1.0/real2nd.T[0], 1.0/real2nd.T[0]**2, ls='dashed', label="$1/n^2$")

plt.xscale("log")
plt.yscale("log")

plt.legend()

plt.title("$n$-dependence of ${\\rm Re}\\lambda$ (simplicial)")
plt.xlabel("$n$ (refinement level)")
# plt.ylabel()

plt.savefig("Relambda_n.pdf")
```
